<h2> What is a Colour Space? </h2>

<i>A color space is a specific organization of colors which alows for reproducible representation of colors in both digital and analog representations.
Color is usually measured by the following:
<p>(1) Brightness: This is the human sensation by which an
area exhibits more or less light.
<p>(2) Hue: This is the human sensation according to which an
area appears to be similar to one, or to proportions of two,
of the perceived colors red, yellow, green and blue.
<p>(3) Colorfulness: This is the human sensation according to
which an area appears to exhibit more or less of its hue.
<p>(4) Lightness: This can be defined as the sensation of an
areaâ€˜s brightness relative to a reference white in the scene.
<p>(5) Chroma: This is the colorfulness of an area relative to
the brightness of a reference white.
<p>(6) Saturation: This is the colorfulness of an area relative
to its brightness.</i>[1]

<h2>What is HSV?</h2>
Another way to represent a color is in terms of the HSV model. The Hue is referred to the color it resembles, Saturation describes how white the color is and the value is represented as the lightness or darkness of the color and is represented in this color cylinder.[3]

![alt text](hsv1.jpg "HSV Color Space")

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

# Read in carotid image
carotid_img = cv2.imread("Carotid.jpg")
                         
# Turn the image into HSV colour space
HSV = cv2.cvtColor(carotid_img, cv2.COLOR_BGR2HSV)
# Convert the image from BGR to RGB
HSV = cv2.cvtColor(HSV, cv2.COLOR_BGR2RGB)

# Show Images
fig = plt.figure(figsize=(15,20))
ax1 = fig.add_subplot(2,2,1)
imgplot = plt.imshow(carotid_img)
ax1.set_title("Original")

ax2 = fig.add_subplot(222)
imgplot2 = plt.imshow(HSV)
ax2.set_title("HSV")
plt.show()

HSV = cv2.cvtColor(HSV, cv2.COLOR_RGB2BGR)
```

<h2> It worked ! Now Let's mask it !</h2>
Using the HSV colour space we were able to pull out the desired areas and its just a matter of cleaning it up.

```python
# Create threshold boundaries
HSVlower = np.array([0,5,130])
HSVupper = np.array([255,255,255])

# Define a mask
mask = cv2.inRange(HSV, HSVlower, HSVupper)

# Show Images
fig = plt.figure(figsize=(15,20))
ax1 = fig.add_subplot(2,2,1)
imgplot = plt.imshow(mask)
ax1.set_title("HSV Mask")
plt.show()
```

<h3> Apply the mask to the original image</h3>
Just to test my mask against the original image

```python
# Applying the mask on the original carotid image
result = cv2.bitwise_and(carotid_img,carotid_img,mask=mask)
result = cv2.cvtColor(result,cv2.COLOR_BGR2RGB)

# Show Images
fig = plt.figure(figsize=(15,20))
ax1 = fig.add_subplot(2,2,1)
imgplot = plt.imshow(result)
ax1.set_title("Result")
plt.show()
```
